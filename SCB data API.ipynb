```python

scb_data = scb.get_data()
scb_data
```

```python
#convert json data to a pandas DataFrame

import pandas as pd
data = scb_data

df = pd.DataFrame([
    {
        'Year': int(item['key'][1]),  # Year is at index 1 in key
        'Number_of_Cars': int(item['values'][0])
    }
    for item in data['data']
])

# Display result
print(df)
result = df.to_excel(r'D:\EC utbildning\R\scb_data.xlsx', index=False)
```

```python
# Plotting the data personal cars in traffic over time
import matplotlib.pyplot as plt
df = df.sort_values('Year')

# Plot
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Number_of_Cars'], marker='o', color='blue')

# Ensure all years are shown on x-axis
plt.xticks(ticks=df['Year'], labels=df['Year'], rotation=45)

plt.title("Number of Personal Cars in Traffic")
plt.xlabel("Year", fontsize=16)
plt.ylabel("Number of Cars", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
